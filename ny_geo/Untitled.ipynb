```python
from geopy.geocoders import Nominatim

# Initialize the geocoder
geolocator = Nominatim(user_agent="my_geocoder")

# Address to geocode
address = "New York City, NY"

# Geocode the address
location = geolocator.geocode(address)

# Print latitude and longitude
print("Latitude:", location.latitude)
print("Longitude:", location.longitude)
```

```python

def text_file_content(file_path):
    with open(file_path, 'r') as file:
        content = file.read().splitlines()
    return(content)

        

# Example usage:
file_path = 'nyc_rentals_ret.txt'  # Replace with the path to your actual text file
text_lines = (text_file_content(file_path))
#print(text)
# Define a regular expression pattern to match addresses
import re

def filter_address_lines(text_lines):
    address_list = []
    address_pattern = re.compile(r'\b\d+\s+(?:\w+\s+){1,2}'
                                 r'(?:St|Ave|Dr|Broadway|Blvd|Americas|Ct|Pl|Rd|Rd W'
                                 r'|Hwy|N|Bay Rd|W|A|E|Pky|Sq|Ln|West)\b'
                                 , re.IGNORECASE)
    for line in text_lines:
        if address_pattern.search(line) or 'Broadway' in line:
            address_list.append(line)
    return address_list


# Example usage:
address = filter_address_lines(text_lines)
print(address)
print(len(address))
```

```python
def get_text_loc(text_lines_list, address_lines_list):
    k=0
    subl_len = []
    for line in text_lines_list:
        k+=1
        if line in address_lines_list:
            subl_len.append(k)
    subl_len = [el - 1 for el in subl_len]
    return subl_len
            
address_pos = (get_text_loc(text_lines, address))
print(address_pos)
print(len(address_pos))
```

```python
def subtract_previous(lst):
    result = []
    for i in range(1, len(lst)):
        result.append(lst[i] - lst[i - 1])
    return result

chunks = (subtract_previous(address_pos))

from itertools import islice

def divide_into_chunks(input_list, chunks_list):
    it = iter(input_list)
    sliced =[list(islice(it, 0, i)) for i in chunks_list]
    return sliced

text_sublists = divide_into_chunks(text_lines[1:], chunks)
print(text_sublists)

```

```python
def is_substring_present(substring, nested_list):
    for sublist in nested_list:
        for text in sublist:
            if substring in text:
                return True
    return False

# Example usage
substring = "example"
nested_list = [["This is an example text", "Another example"], ["Not this one"]]

is_substring_present(substring, nested_list)
```
